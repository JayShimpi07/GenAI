TASK 4: ChatPromptTemplate System

```python
def get_chat_prompt(role, topic): # Define a function that takes a role and a topic as input and returns a formatted chat prompt
    if role == "teacher": # Check if the role is "teacher"
        system_msg = "You are a helpful teacher." # If the role is "teacher", set the system message to "You are a helpful teacher."
    elif role == "interviewer": # Check if the role is "interviewer"
        system_msg = "You are an interviewer asking questions." # If the role is "interviewer", set the system message to "You are an interviewer asking questions."
    elif role == "motivator": # Check if the role is "motivator"
        system_msg = "You are a motivational speaker." # If the role is "motivator", set the system message to "You are a motivational speaker."
    else:
        system_msg = "You are a helpful assistant." # For any other role, set the system message to "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([ # Create a ChatPromptTemplate instance from a list of messages
        ("system", system_msg), # Add the system message to the chat template
        ("user", "Explain {topic}") # Add a user message with a placeholder for the "topic" variable to the chat template
    ])

    return chat_template.format_messages(topic=topic) # Format the chat template with the topic variable and return the formatted messages


print("Task 4 Output:")
messages = get_chat_prompt("teacher", "Neural Networks") # Call the get_chat_prompt function with the role "teacher" and topic "Neural Networks" and store the returned messages in the variable "messages"
for msg in messages: # Iterate over the messages in the "messages" variable
    print(msg)
print("\n")
```

TASK 5: Input Validation Layer

```python
def validate_inputs(audience, tone): # Define a function that takes an audience and a tone as input and validates them against predefined lists of valid audiences and tones
    valid_audience = ["beginner", "intermediate", "expert"] # Define a list of valid audiences
    valid_tone = ["formal", "casual", "fun"] # Define a list of valid tones

    if audience not in valid_audience: # Check if the input audience is not in the list of valid audiences
        print("Invalid audience! Defaulting to 'beginner'") # If the audience is invalid, print a message and default to "beginner"
        audience = "beginner" # Set the audience variable to "beginner"

    if tone not in valid_tone: # Check if the input tone is not in the list of valid tones
        print("Invalid tone! Defaulting to 'formal'") # If the tone is invalid, print a message and default to "formal"
        tone = "formal" # Set the tone variable to "formal"

    return audience, tone # Return the validated audience and tone


# Test
print("Task 5 Output:")
aud, tone = validate_inputs("kids", "friendly") # Call the validate_inputs function with the audience "kids" and tone "friendly" and store the returned validated audience and tone in the variables "aud" and "tone"
print(aud, tone)
print("\n")
```

TASK 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style): # Define a function that takes a topic, audience, tone, and style as input and generates a formatted prompt based on the style
    
    # Validate input
    audience, tone = validate_inputs(audience, tone) # Call the validate_inputs function with the input audience and tone to validate them and store the validated audience and tone in the variables "audience" and "tone"

    if style == "teaching": # Check if the style is "teaching"
        template = PromptTemplate( # Create a PromptTemplate instance for teaching style
            input_variables=["topic", "audience", "tone"], # Define the input variables "topic", "audience", and "tone" that will be used in the template
            template="Explain {topic} for {audience} in a {tone} teaching style" # Define the template string with placeholders for the "topic", "audience", and "tone" variables
        )

    elif style == "interview": # Check if the style is "interview"
        template = PromptTemplate( # Create a PromptTemplate instance for interview style
            input_variables=["topic", "audience", "tone"], # Define the input variables "topic", "audience", and "tone" that will be used in the template
            template="Ask interview questions about {topic} for {audience} in a {tone} tone" # Define the template string with placeholders for the "topic", "audience", and "tone" variables
        )

    elif style == "storytelling": # Check if the style is "storytelling"
        template = PromptTemplate( # Create a PromptTemplate instance for storytelling style
            input_variables=["topic", "audience", "tone"], # Define the input variables "topic", "audience", and "tone" that will be used in the template
            template="Explain {topic} for {audience} in a {tone} storytelling style" # Define the template string with placeholders for the "topic", "audience", and "tone" variables
        )

    else:
        return "Invalid style"

    return template.format(topic=topic, audience=audience, tone=tone) # Format the selected template with the topic, validated audience, and validated tone and return the formatted prompt


print("Task 6 Output:")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling")) # Call the generate_prompt function with the topic "Neural Networks", audience "beginner", tone "fun", and style "storytelling" and print the generated prompt
print("\n")
```

TASK 7: Template Reusability Test

```python
reuse_template = PromptTemplate( # Create a PromptTemplate instance for reuse
    input_variables=["topic", "audience", "tone"], # Define the input variables "topic", "audience", and "tone" that will be used in the template
    template="Explain {topic} for {audience} in a {tone} tone" # Define the template string with placeholders for the "topic", "audience", and "tone" variables
)

inputs = [ # Define a list of input tuples containing topic, audience, and tone for testing the reuse of the template
    ("AI", "beginner", "fun"),
    ("Python", "expert", "formal"),
    ("Data Science", "intermediate", "casual"),
    ("Cyber Security", "beginner", "formal"),
    ("Machine Learning", "expert", "technical")
]

print("Task 7 Outputs:")
for t, a, to in inputs: # Iterate over the input tuples in the "inputs" list and unpack each tuple into the variables "t", "a", and "to" for topic, audience, and tone respectively
    print(reuse_template.format(topic=t, audience=a, tone=to)) # Format the reuse_template with the topic, audience, and tone from the current tuple and print the output
```
